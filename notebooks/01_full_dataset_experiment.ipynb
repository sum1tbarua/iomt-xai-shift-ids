# Import Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
```

# Dataset Load

```python
train_df = pd.read_csv("datasets/CICIOMT/train/ciciomt2024_train_balanced.csv")
val_df = pd.read_csv("datasets/CICIOMT/val/ciciomt2024_validation_balanced.csv")
test_df = pd.read_csv("datasets/CICIOMT/test/ciciomt2024_test_balanced.csv")
```

# Shape and Label Distribution

```python
print("Training Data Shape: ", train_df.shape)
print("Validation Data Shape: ", val_df.shape)
print("Testing Data Shape: ", test_df.shape)

# Label distribution in training data
print("Label Distribution in Training Data:")
print(train_df['binary_label'].value_counts())

# Label distribution in validation data
print("\nLabel Distribution in Validation Data:")
print(val_df['binary_label'].value_counts())

# Label distribution in testing data
print("\nLabel Distribution in Testing Data:")
print(test_df['binary_label'].value_counts())
```

# Define feature columns and target label

```python
non_feature_cols = [
    "source_file",
    "attack_type",
    "binary_label"
]

feature_cols = [col for col in train_df.columns if col not in non_feature_cols]
target_label = "binary_label"
```

# Define train, validation and test sets

```python
# Creating features for training
X_train = train_df[feature_cols]
y_train = train_df[target_label]

# Creating features for validation
X_val = val_df[feature_cols]
y_val = val_df[target_label]

# Creating features for test
X_test = test_df[feature_cols]
y_test = test_df[target_label]
```

# Model Training

## Logistic Regression

```python
# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

# logistic regression model
logreg = LogisticRegression(
    max_iter=1000,
    random_state=42,
    n_jobs=-1
)

# train
logreg.fit(X_train_scaled, y_train)

print("Validation Metrics")
# prediction
y_val_pred = logreg.predict(X_val_scaled)
y_val_prob = logreg.predict_proba(X_val_scaled)[:, 1]

# metrics
print("Accuracy:", accuracy_score(y_val, y_val_pred))
print("Precision:", precision_score(y_val, y_val_pred))
print("Recall:", recall_score(y_val, y_val_pred))
print("F1:", f1_score(y_val, y_val_pred))
print("ROC-AUC:", roc_auc_score(y_val, y_val_prob))

print("Confusion Matrix:")
print(confusion_matrix(y_val, y_val_pred))


print("\n\nTest Metrics")
X_test_scaled = scaler.transform(X_test)

# Predictions on test set
y_test_pred = logreg.predict(X_test_scaled)
y_test_prob = logreg.predict_proba(X_test_scaled)[:, 1]

# Metrics on test set
print("Logistic Regression Test Set Metrics:")
print("Accuracy:", accuracy_score(y_test, y_test_pred))
print("Precision:", precision_score(y_test, y_test_pred))
print("Recall:", recall_score(y_test, y_test_pred))
print("F1:", f1_score(y_test, y_test_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred))
```

## Random Forest

```python
# Random Forest model
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

# train
rf.fit(X_train, y_train)

print("Validation Metrics")
# predictions
y_val_pred = rf.predict(X_val)
y_val_prob = rf.predict_proba(X_val)[:, 1]

# metrics
print("Accuracy:", accuracy_score(y_val, y_val_pred))
print("Precision:", precision_score(y_val, y_val_pred))
print("Recall:", recall_score(y_val, y_val_pred))
print("F1:", f1_score(y_val, y_val_pred))
print("ROC-AUC:", roc_auc_score(y_val, y_val_prob))

print("Confusion Matrix:")
print(confusion_matrix(y_val, y_val_pred))

print("\n\nTest Metrics")
# Testing RF
y_test_pred = rf.predict(X_test)
y_test_prob = rf.predict_proba(X_test)[:,1]

print("Random Forest Test Set Metrics:")
print("Accuracy:", accuracy_score(y_test, y_test_pred))
print("Precision:", precision_score(y_test, y_test_pred))
print("Recall:", recall_score(y_test, y_test_pred))
print("F1:", f1_score(y_test, y_test_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred))
```

## XGBoost

```python
# XGBoost model
xgb = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1,
    eval_metric='logloss'
)

# train
xgb.fit(X_train, y_train)

print("Validation Metrics")
y_val_pred = xgb.predict(X_val)
y_val_prob = xgb.predict_proba(X_val)[:, 1]

print("Accuracy:", accuracy_score(y_val, y_val_pred))
print("Precision:", precision_score(y_val, y_val_pred))
print("Recall:", recall_score(y_val, y_val_pred))
print("F1:", f1_score(y_val, y_val_pred))
print("ROC-AUC:", roc_auc_score(y_val, y_val_prob))

print("Confusion Matrix:")
print(confusion_matrix(y_val, y_val_pred))

print("\n\nTest Metrics")
# Test on XGBoost
y_test_pred = xgb.predict(X_test)
y_test_prob = xgb.predict_proba(X_test)[:, 1]

print("XGBoost Test Set Metrics:")
print("Accuracy:", accuracy_score(y_test, y_test_pred))
print("Precision:", precision_score(y_test, y_test_pred))
print("Recall:", recall_score(y_test, y_test_pred))
print("F1:", f1_score(y_test, y_test_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred))
```
